Artificial intelligence allows us to access knowledge that we as individuals could never hope to analyze and read ourselves. But this raises some problems that we aren’t used to thinking about. When do we have too much knowledge? How much should we know, as individuals and corporations and governments? Who should know what when?

* **Humans Create Data.** Many corporations now tout their data capital as one of their major assets. In other words, the value of these corporations is largely derived from the data that they claim to own. This raises an important question for us: who does own language data? Intellectual property has always been a commercial asset. For example, a publishing house controls the text of the books they publish. The difference is that the publishing house pays the author of the book for that right. How do tech companies like Google and Facebook pay the authors of the data that they rely on? The answer is clearly that they do not pay their users in the same way that publishers pay their authors. But the point here is that, by definition, humans create the language data that we need for AI.

* **Data Creates AI.** We’ve seen already that AI needs written language to learn about the world. Without training data, machine learning is powerless. What this means is that AI is the offspring of two parents: human data and open source scientific software (see below). And here’s the question: If data is a tradable asset, and if AI is a direct output of that data, does this mean that the humans who author the training data maintain ownership over a fully trained AI model? In other words, if everyone shares in the creation of AI (through creating written knowledge), does everyone share in the ownership of that AI?

* **Open Source Software Creates AI.** The second parent of AI is software. It turns out that most of the scientific software the makes AI possible is open source: created by scientists and enthusiasts for the benefit of society at large. Tech companies like Google and Microsoft benefit from this software. To what extend to they contribute? As shown in the figure below, it turns out that a lot of tech companies contribute back to the ecosystem they depend on. This figure is showing the number of active contributors to public GitHub repositories as of mid 2020. The higher the number of contributions (here the more accounts are making public commits), the more each company is giving back open source software to the wider community. A commit in this context is a single act of adding a contribution. Of course, this figure hasn’t been adjusted to reflect the total size and value of each company. Red Hat, for example, is giving back far more as a proportion of its total size. Facebook and Amazon are often seen as exploiters rather than contributors because of their low number of open source contributions.

* **Who will benefit?** So the basic situation is that (1) everyday humans like you and me create the data and (2) open source coders and scientists and companies (the ones who actually give back) create the technology that makes AI possible. But who benefits from this work? This a relatively new area for human thought, and so our thinking about the ethics of AI hasn’t caught up with the reality of AI. But an analogy helps put this into context: imagine a community that comes together to plant a community garden. Some citizens brings in the seeds and the fertilizer (data). And others spend their weekends weeding and cultivating the garden itself (open source software). At harvest time, the owner of a local grocery store (like Amazon) comes in the night and picks all the crops. Then sells them back to the local residents. Is this act of harvesting right or wrong?

* **Could there be privacy concerns?** AI only learns what we train it to learn. If the training data contains individual identifying information that creates privacy concerns, the model inherits this breach. Here’s another story: imagine we train an authorship model to verify that a bank customer who tries to access an account is actually the owner of that account. The bank has millions of customers in the US: mostly speakers from America, Canada, Mexico, and the UK. Now, New Zealand has a unique dialect and only a handful of the bank’s data comes from New Zealanders. Then there’s a grisly murder. The perpetrator withdrew money from an ATM near the scene. The bank hands over the authorship model’s records to the police: our model says it was a speaker from New Zealand! But the authorship model was trained and evaluated to distinguish between millions of non-New Zealanders. Do we trust the model to remain accurate for an unseen minority class? Do the authorities who are now relying on that model even understand the skewed data that it was trained with?

* **Could this enable discrimination?** AI only knows what it finds in the data we train it with. So, if we train a model using data that implies that Mexicans are lazy, African-Americans are violent, Irish Catholics are drunks, and Polish people are unintelligent… then AI will learn that these slurs are facts. Our treasure chest of human knowledge contains our best thinkers as well as our worst thoughts. And machine learning has no idea which is which. Should we care whether our models are racist and xenophobic? Would this bias disqualify an otherwise accurate model?

* **Governments.** Corporations use data and models for profit, as assets. The other major player in this field is governments. Nation states tend to use text analytics for four purposes: (1) for forensic investigations in criminal cases, (2) for surveillance and spycraft on domestic individuals, (3) for surveillance and spycraft on foreign individuals, (4) for large-scale digital influence or propaganda campaigns. How can we think about these governmental activities from an ethical perspective?

>* First, some countries (like the US) take a position that makes domestic surveillance illegal while continuing to focus on citizens of other countries. And other countries (like Russia and China) take a position that accepts surveillance on their own domestic citizens (as an issue for internal security). So one ethical position is that the use of the technology itself is not problematic: the ethical issue is about the target. For some countries internal domestic surveillance is a sovereign right and for others external foreign surveillance is a matter of national defense.

>* Second, to what degree do naïve humans understand how much data they are producing? It is fair to say that most individuals do not understand the implications of carrying a location-tracking device (their smart phone). And it’s fair to say that most individuals do not understand the access that some apps (like Facebook) give third-parties to their user data and text. And here’s the point of asking this question: one approach is to say that the ethical question comes down to who is the target. But most people don’t understand the degree to which they have become targets themselves. Is awareness and consent required when the ownership of data is transferred? This is a question both for selling data (corporations) and for intercepting data (governments).

>* Third, all of these applications, like surveillance and propaganda, have existed long before the birth of AI. Is AI-driven propaganda any different from traditional propaganda? Is AI-driven surveillance any different from traditional surveillance? The question here is whether the scale that AI supports changes the underlying ethics that are involved.

>* Fourth, is the technology problematic, or is the problem with specific applications of the technology? For example, the same technology that supports misinformation during an election (targeting individuals based on their behavior with specific messages) could help spread proper food safety techniques or disaster response information (saving thousands of lives in the process). Do the benefits of good applications outweigh the harms of bad applications? Is there even a defendable distinction between propaganda (influencing an election) and a public awareness campaign (influencing public opinion)?
