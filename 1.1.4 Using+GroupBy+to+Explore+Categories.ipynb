<h2>The Data Sources</h2>

**We have to train AI, using different kinds of documents to learn different types of knowledge.**

Humans don’t write down obvious information. For example, physics has many more men than women, with one of the largest gender gaps of any academic field. But the phrase female physicist is actually more common than male physicist. Humans never mention that kind of obvious default information. Another example: some of the most common adjectives used with boss are “good” and “political” and “female.” This tells us that our default image of a boss is a poorly performing man in a business setting. The most common adjectives for pizza are “frozen” and “free” and “cold.” Again, our typical experience is that we pay for freshly made pizza that is still hot. My point is that only unusual or unexpected properties are explicitly stated, which means that AI is biased toward whatever humans have decided is worth mentioning.


1.4_Pizza.jpg

This makes implicit knowledge difficult for AI to learn. Think about the metaphor, “I’ve been putting some money aside.” This is part of a network of metaphors that views money as a form of storage: for example, “Workers need more money in their pockets right now.” Humans often use metaphors like this that are indirect, saying something that they don’t literally mean. Now think about a joke: “It is so cold outside I saw a politician with his hands in his own pockets.” The phrase hands in his own pockets is ambiguous, has two different meanings. It depends on that metaphor to create a humorous contrast: his hands are usually not in his own pockets because he isn’t cold, but also because they’re in someone else’s pockets! Much of the meaning in the documents that we want to analyze is actually implicit, indirect, unstated. Some of it is non-literal but nonetheless true (like the metaphor) and some of it is half literal and partially true (like the joke).

Remember that our goal is to use text analytics to make sense out of written documents. But it turns out that our first problem is to teach or train artificial intelligence about the world. If our models don’t have a basic understanding of how the world works, then they can’t find anything useful in written documents. AI has the advantage that texts provide more explicit knowledge than any individual human could possibly know. So we can train our models using more raw facts than a human has access to. But what about the background context that makes these facts meaningful? What about the implicit knowledge that humans never bother to write down?

Where we collect data depends on the knowledge that we want to learn.

* News articles are a good source for changing factual information. But news articles usually don’t explain the background information we need to understand what is changing. For example, an article about American elections assumes that we know some countries choose leaders through voting, that some people can vote and others can’t, that only one vote is cast per person, etc. But none of this is obvious if you don’t already know it.

* Wikipedia is a good source for background factual information. A Wikipedia article won’t tell us what just happened, only basic information about how the world works. One of the reasons that Wikipedia is such a great resource is that its articles are linked together and categorized. Humans might know that Doc Martins are a kind of shoe and that shoes are covering for your feet. But AI would never know this if there wasn’t a resource like Wikipedia that systematically linked and catalogued that kind of knowledge.

* Social media is a good source for information about everyday life. News articles and Wikipedia only tell us about a small number of events from a small number of perspectives. Most people never meet their Prime Minister and most daily events have nothing to do with Parliament. This kind of every day information just isn’t contained in news articles. A lot of people wear shoes everyday, but newspapers and Wikipedia won’t help us learn anything about what it is like to actually wear them or why we wear different types of shoes in different settings.

* Novels and other literature are a good source for implicit social information. We as humans often aren’t very aware of our own assumptions or just how our societies work. And, if humans don’t realize something, they can’t put that knowledge explicitly into texts.**bold text**

# Lecture Lab

In this lab, we are going to further explore categories in our dataset. We're going to do this by looking at one category or class at a time. So, we're going to look at news article. But instead of seeing them all at once, we're going to look at one year at a time.

We start by loading our environment.

```
from text_analytics import text_analytics
import os
import pandas as pd

ai = text_analytics()
print("Done!")
```

Now we're going to load our data. This time it will be articles from *Politico*.

```
file =os.path.join(ai.data_dir, "Wordclouds.Politico.gz")
#quick hack to add a leading "/" to the path

df = pd.read_csv(file, index_col = 0)
print(df)
```

So, here is how we look at one category at a time: using *groupby*.

```
for year, year_df in df.groupby("Year"):
        print(year, ": ", len(year_df))
        
print("Done!")
```

This code shows us the number of articles per year that the dataset has. This is one way of iterating over the data. Except that instead of iterating over individual articles we are iterating over categories of articles.

Let's turn this into a graph now.

```
years = []
for year, year_df in df.groupby("Year"):
        years.append([year, len(year_df)])

years = pd.DataFrame(years, columns = ["Year", "N."])
years.set_index("Year", drop = True, inplace = True)
print(years)
print("Done!")
```

```
years.plot(kind = "barh")
print("Done!")
```
